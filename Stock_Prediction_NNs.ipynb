# Import Data From YFinance

```python
import yfinance
import numpy as np
from tensorflow import keras
```

```python
microsoft = yfinance.Ticker("MSFT")
data = microsoft.history()

y_train = data.Close
print(y_train)
x_train = np.arange(0, len(y_train))
print(x_train)
```

# Building Neural Network

```python
from keras.src.layers.serialization import activation

mlp = keras.Sequential([
    keras.layers.Input(shape = 1),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(1)
])

adamOptimizer = keras.optimizers.Adam()
mlp.compile(optimizer=adamOptimizer, loss="mean_squared_error")

mlp.fit(x_train, y_train, epochs = 5000, validation_split=.1)

mlp.predict([1])
```
